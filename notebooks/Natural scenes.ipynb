## Motivation

1. **Characterizing receptive fields** - A fundamental goal of sensory neuroscience is to recover the transformations performed by early sensory neurons on incoming stimuli.  Consequently, a substantial component of research in this area has focused on methods for characterizing the receptive fields of sensory neurons.  One recent approach that has been widely adopted is presenting white (Gaussian) noise stimuli to cells.  This approach offers numerous advantages over traditional methods, one of which is the toolbox conferred by linear systems analysis.  Methods such as Covariance analysis can be readily applied to the ensemble of spikes (Spike Triggered Ensemble, STE) generated by a noisy stimulus to recover higher order structure.  
<br>
2. **Advantages of white noise** - The traditional approach to characterizing early sensory neurons is to present highly parameterized stimuli, such as spots, bars, step currents or tones.  This approach has proven highly effective in characterizing linear models of neural response.  However, these approaches introduce significant disadvantages, such as limited stimulus range, potential adaptation confounds and difficulty extracting non-linear components (Chichilnisky, 2000).  The stochastic approach of white noise analysis can complement and extend these classic methods to characterize neurons with arbitrarily complex response properties.  White noise is stochastic, avoiding adaptation, widely varying, permitting a wider range of stimulus space to be explored, and capable of recovering non-linearities in the response properties of neurons.
<br>
<br>
3. **Natural statistics** - A commonly cited downside of both methods of neural characterization described above is the discrepancy between the artificial stimuli and the known characteristics of natural scene or auditory statistics.  The first and second moments of natural stimuli have been extensively studied over the past three decades and, in comparison to white noise, are known to exhibit significant structure.  In natural scenes, for example, long range correlations obeying a roughly $1/f^2$ (where $f$ represents frequency) power law have been described in both the temporal and spatial domains.  These observations have motivated interest in studying how the nervous system might take advantage of this statistical regularity, a question that necessitates the use of more complex stimuli.  Such stimuli can often expose response regimes that are rarely found in a rapidly varying white stimulus.

## Goals

1. **Implement Spike Triggered Covariance (STC) on correlated stimuli** - While some authors have begun presenting natural videos and sounds to sensory neurons, we take an intermediate approach and construct temporally correlated noise.  These stimuli will lack the phase structure of natural stimuli, but will provide a more naturalistic power structure.  Regardless, the analysis outlined here closely follows an approach used to study visual neurons response to natural images (Touryan et al. 2005) and should therefore be easily extended to natural stimuli.  Our first goal is simply to implement these analyses, albeit using a more constructed stimulus set.
<br>
<br>
2. **Compare receptive field estimates** - In addition to implementing the analysis, the second goal is to compare the estimated receptive fields using noise with differing degrees of temporal correlation.  However, for simplicity we chose a biophysical model that is deterministic and contains only 3 ionic currents.  We therefore do not expect substantial variations in receptive field after accounting for correlations in the stimuli as the receptive field of this model neuron is known.  Therefore, the comparison will serve as more of a test of the efficacy of receptive field recovery.

## Approach

There have been numerous approaches proposed for estimating receptive fields using complex stimuli, such as natural scenes.  These include random least squares methods, neural networks, information maximization and STC (Felson &amp; Dan, 2005).  We use chose to use the STC method because it is generally faster and easier to implement than the other approaches.  The STC approach requires a correction for the correlations in the stimuli.  We use the method described by Touryan et al. (2005). We describe this in more detail below.

### I. Constructing stimuli
***

The goal of this project was to simulate the effects of introducing correlation into Gaussian stimuli during noise analysis.  The function defined below, `CorrelatedNoise1D`, introduces correlation to a sequence of random numbers in the Fourier domain.  

The steps taken are as follows:

1. The function begins by finding the next power of two, which will speed up computations in the Fourier transform later. 
2. The noise sequence is then generated using a Mersenne twister algorithm.  As an optional parameter, the user can specify a seed for the algorithm that can be useful when regenerating the stimuli later or for the purposes of comparison.  Here, we pass in a seed in order to directly compare the effect of correlation on neuronal responses.  Using the exact same random sequence for each stimulus will eliminate random variability due to finite sampling.   
3. The noise sequence is then moved into the Fourier domain, where convolution between a signal and a filter (i.e. $1/f^{\alpha}$) can be computed much more efficiently through multiplication.  
4. We then generate a vector, n, with the $1/f^{\alpha}$ properties specified by the parameter `alphaValue`.  An $\alpha = 0$ will not introduce any correlation (anything raised to 0 equals 1), while increasing $\alpha$ increases the correlation in the signal.  Since we are manipulating the power spectrum, we take the square root of n.
5. The left hand of the Fourier spectrum is multiplied by the power spectrum. Since the Fourier spectrum is symmetric, we flip the power spectrum, n, and then multiply the second half.
6. The inverse Fourier transform is than taken to move the signal back into the time domain and we select the first half of the signal (since it is symmetric).
7. Finally, the signal is corrected to have the correct mean and standard deviation.


```python
def CorrelatedNoise1D(numSamples, standardDev=0.3, meanValue=0, alphaValue=1, seed=None):
    """
    Produces $\frac{1}{f^\alpha}$ temporally correlated noise.

    :param numSamples: number of samples to generate
    :param standardDev: standard deviation of the noise (default = 0.3)
    :param meanValue: mean value of the noise (default = 0);
    :param alphaValue: type of correlation (i.e., 1 = 1/f, 2 = 1/f^2)

    :returns: y = row vector of the correlated noise
    """

    # 1. Find the next power of two.
    nextP2 = 2 ** np.ceil(np.log2(abs(numSamples)))
    
    # 2. Generate the noise sequence.
    if seed != None:
        np.random.seed(seed)
    x = np.random.rand(nextP2)
    
    # 3. calculate fft and then perform 1/f multiplication
    f = np.fft.fft(x)
    
    # 4. prepare a vector for 1/f^alpha multiplication
    n = np.arange(1, nextP2 / 2 + 2) ** alphaValue
    # It's the power spectrum, so take the square root.
    n = np.sqrt(n) 

    # 5. Multiply the left half of the fft by the power spectrum
    f[:nextP2 / 2 + 1] = f[:nextP2 / 2 + 1] / n
    f[nextP2 / 2 - 1:] = f[nextP2 / 2 - 1:] / n[::-1]
   
    # 6. Take the inverse FFT.
    y = np.fft.ifft(f) # 
    y = np.real(y[:numSamples]) # Just take the real component.
    y = y.T # Make sure it's a row.
    
    # 7. Make sure it's the right mean and standard deviation.
    y = y / np.std(y) * standardDev
    y = y - np.mean(y) + meanValue

    return y
```

### II. $1 / f^{\alpha}$ noise: manipulating $\alpha$
***

Below we generate correlated noise with three different $\alpha$ values.  For this experiment we chose 0.5, 1, and 2.  All other parameters (mean, standard deviation and seed) are identical in order to facilitate direct analysis of correlation.  We did not use an $\alpha$ of 0 because a perfectly random signal will rarely produce a spike as it moves too quickly between positive and negative currents.  At least a small amount of correlation is necessary to properly characterize the cell.

```python
# Generate noise stimuli with various degrees of correlation:
stimA = CorrelatedNoise1D(tim / dt, 5, 0, 0.5, 1)
stimB = CorrelatedNoise1D(tim / dt, 5, 0, 1.0, 1)
stimC = CorrelatedNoise1D(tim / dt, 5, 0, 2.0, 1)

# Compute power spectrum:
powerA = abs(np.fft.fftshift(np.real(np.fft.fft(stimA ** 2))))
powerB = abs(np.fft.fftshift(np.real(np.fft.fft(stimB ** 2))))
powerC = abs(np.fft.fftshift(np.real(np.fft.fft(stimC ** 2))))
```

### III. Hodgkin-Huxley model
***

In 1952, Hodgkin and Huxley experimentally derived a set of equations that described the initiation and propagation of an action potential. The primary equation, describing a neuron's membrane potential, is as follows: 

$$C_{m} \frac{dV_{m}}{dt} = I_{ext} - g_{K}n^4(V_{m}-E_{K}) - g_{Na}m^3h(V_{m}-ENa) - g_{l}(V_{m}-E_{l})$$

$n$, $m$, and $h$ are the gating variable that correspond to the probability that a channel is open bound between [0,1], and are described by a set of differential equations,

$$\frac{dm}{dt} = \alpha_{m}(V)(1-m)-\beta_{m}(V)m$$
<br>
$$\frac{dn}{dt} = \alpha_{n}(V)(1-n)-\beta_{n}(V)n$$
<br>
$$\frac{dh}{dt} = \alpha_{h}(V)(1-h)-\beta_{h}(V)h$$

The rate constants $\alpha$ and $\beta$ of each equation, are themselves described by a set of voltage dependent differential equations. Below, we use these equations, as well as the resting (initial) values, to simulate a neuron firing an action potential.

We adopt here the parameters used by Agüera y Arcas et al. (2003), which are the original parameters with a sign flip to conform to modern conventions of reporting voltage.

```python
def HHmodel(Current, dt):
    """Integrates the Hodgkin Huxley equations.
    The current parameters of the model are taken from 
    """
    TIME = len(Current) # in msec
    dt = dt # in msec
    
    # Define the cell's initial conditions
    # Conductances (in mS/cm^2):
    gK = 36
    gNa = 120
    gL = 0.3
    
    # Define the equilibrium potentials and conductances for sodium, potassium,
    # and the leak current
    VK = -12
    VNa = 115
    VL = 10.613

    V = 0 ##mV
    n = 0.0816
    m = 0.1003
    h = 0.0643
    
    # Functions to Calculate alphas and betas for the intial membrane voltage (based on
    #the equations described by Hodgkin and Huxley, modified to refelct the
    # initial Vm)
    
    # K channel
    alpha_n = np.vectorize(lambda v: 0.01 * (-v + 10) / 
                (np.exp((-v + 10) / 10) - 1) if v != 10 else 0.1)
    beta_n  = lambda v: 0.125 * np.exp(-v / 80)
    
    # Na channel (activating)
    alpha_m = np.vectorize(lambda v: 0.1 * (-v + 25) / 
                (np.exp((-v + 25) / 10) - 1) if v != 25 else 1)
    beta_m  = lambda v: 4 * np.exp(-v / 18)
    
    # Na channel (inactivating)
    alpha_h = lambda v: 0.07 * np.exp(-v / 20)
    beta_h  = lambda v: 1 / (np.exp((-v + 30) / 10) + 1)
    
    
    ## Preallocate for speed:
    Voltage = np.zeros((TIME))
    Voltage[0] = V
    
    # Euler integration of the equations
    for i in range(0, TIME):
    
        I = Current[i]
        
        # The equations that describe the gating variables:
        m += dt * (alpha_m(V) * (1 - m) - beta_m(V) * m)
        h += dt * (alpha_h(V) * (1 - h) - beta_h(V) * h)
        n += dt * (alpha_n(V) * (1 - n) - beta_n(V) * n)
    
        # The equation describing the cell's net ionic current, rearranged to
        # solve for Vm.
        V = V + ( (I) - ((gK * (n ** 4))*(V - VK)) - ((gNa * (m ** 3)) 
                * h * (V - VNa)) - (gL * (V - VL))) * dt
        
        Voltage[i] = V
    
    return Voltage
    
```

### IV. Spike Triggered Average (STA)
***

The first step in computing the STA of a neuron is to find the locations of spikes in the data.  Since we are simulating the voltage trace (equivalent to whole cell recording), the detection of spikes is very simple. The function called `findSpikes` below efficiently picks out the spikes with a threshold method:

1. The voltage trace is thresholded at 20mV.  That is, values above 20mV will equal 1, while those less than or equal to 20 will become 0.
2. The thresholded data is then passed to a function called runs which finds the beginning and end of 'runs' of 1s.  This gives us the location of the threshold crossing at the rising and falling phase of the spike.
3. Spike times are checked to ensure that spikes falling too near the beginning or end of the trail are ignored (otherwise they would mess up the STA).

Now having a list of the spike locations, we can extract the current preceding each spike and place it into a matrix which we will call the Spike Triggered Ensemble (STE).  The function `findSTE` will accomplish this task.  This function will extract the current preceding each spike, as well as an equal number of randomly generated spikes that will form the prior distribution (i.e. noise).  The algorithm for this procedure is actually as straightforward as it sounds, as indicated in the code comments.

```python
def runs(bits):
    '''Find zero crossing points from a boolean array.

    :param bits: boolean array.
    '''
    # make sure all runs of ones are well-bounded
    bounded = np.hstack(([0], bits, [0]))
    # get 1 at run starts and -1 at run ends
    difs = np.diff(bounded)
    run_starts, = np.where(difs > 0)
    run_ends, = np.where(difs < 0)

    return run_starts, run_ends

def findSpikes(Voltage, dt):
    '''Find Spikes while ignoring those that fall too close \
    to the beginning or end.

    :param Voltage: voltage trace.
    :param dt: time step used during sampling.

    :returns: Two arrays. One containing location of spike start times \
    the other constainig the spike end times.
    '''
    # 1. Threshold = 20mV
    Spikes = Voltage > 20 
    
    # 2. Find beginning and end of spikes
    Spikes = runs(Spikes)
    
    # 3. Make sure spikes don't fall too close to beginning or end
    # of the trial. 
    S_beg = np.where(Spikes[0] > int(40 / dt))
    S_end = np.where(Spikes[0] < (len(Voltage) - (int(40 / dt))))
    # create an index of spike times
    Length = np.arange(min(S_beg[0]),max(S_end[0]) + 1)
    Spikes = Spikes[0][Length], Spikes[1][Length]
    
    return Spikes

def findSTE(current, Voltage, Spikes, dt, STA_len=40, downS=5):
    '''Find spike triggered ensamble.

    :param current: stimulus played to neuron.
    :param Voltage: voltage trace.
    :param Spikes: a list of spike locations (beginning, end). 
    :param dt: time step used.
    :param STA_len: length of the STA (in msec).

    :returns: Spike triggered ensemble and a randomly generated \
    ensemble.
    '''
    # Preallocate memory for ensamble of spike triggered current:
    STE = np.zeros((len(Spikes[0]), int(STA_len / dt / downS)))
    STE_noise = np.zeros((len(Spikes[0]), int(STA_len / dt / downS)))
    
    # 1. Extract current preceeding spikes:
    for i in range(0,len(Spikes[0])):
        
        # 2. Extract current preceeding spikes:
        Peak = np.arange(Spikes[0][i],Spikes[1][i])
        Peak = Voltage[Peak]
        Height = np.argmax(Peak)
        
        # 3. Find spike location and range:
        Loc = Height + Spikes[0][i]
        # Find a random location from the stimulus:
        RandLoc = np.random.random_integers(STA_len / dt, 
                                            (len(current) - (STA_len / dt)))
        
        Range = np.arange(Loc - (STA_len / dt), Loc, dtype=int)
        RandRange = np.arange(RandLoc - (STA_len / dt), RandLoc, dtype=int)
        
        # 4. Extract currents and downsample if required:
        cur = current[Range]
        STE[i, :] = cur[::downS]
        cur = current[RandRange]
        STE_noise[i, :] = cur[::downS]
    
    return STE, STE_noise

def findFiringRate(curr_Record, tim):
    '''Returns firing rate in spikes/sec.

    :param curr_Record: matrix with current preceeding each spike in rows
    :param tim: time of entire epoch in seconds

    :returns: spike count and spike rate (spikes / second)
    '''
    return len(curr_Record[:, 0]), len(curr_Record[:, 0]) / tim
```

####Compute STA:

Now we find spikes and compile the STE using the above defined functions.  Finally, we print the spike rate for each alpha value and compute the STA, simply taking the mean of across the STE at each sampling position.  Note the large difference in firing rates between the conditions.  This confirms the observation of Touryan et al. (2005) and others who note that introducing correlations into the stimulus increases the firing rate, improving STC analysis (more spikes lead to a better estimate).

```python
# Find spikes in each voltage trace:
spikesA = findSpikes(VoltageA, dt)
spikesB = findSpikes(VoltageB, dt)
spikesC = findSpikes(VoltageC, dt)

# Find STE in each voltage trace & downsample:
STE_A, STE_noiseA = findSTE(stimA, VoltageA, spikesA, dt, STA_len, downS)
STE_B, STE_noiseB = findSTE(stimB, VoltageB, spikesB, dt, STA_len, downS)
STE_C, STE_noiseC = findSTE(stimC, VoltageC, spikesC, dt, STA_len, downS)

# Compute firing rate:
spike_countA, spikeRateA = findFiringRate(STE_A, tim / 1000) # divide by 1000 to get sec
spike_countB, spikeRateB = findFiringRate(STE_B, tim / 1000)
spike_countC, spikeRateC = findFiringRate(STE_C, tim / 1000)

print 'spikes / sec'
print '-------------------'
print 'alpha 0.5:', spikeRateA
print 'alpha 1.0:', spikeRateB
print 'alpha 2.0:', spikeRateC 

# Mean across the first dimension to compute STA:
STA_A = STE_A.mean(0)
STA_B = STE_B.mean(0)
STA_C = STE_C.mean(0)
```

### V. Spike triggered covariance
***

The covariance matrix of the spike triggered ensemble is defined as:

$$C_{spike} = \frac{1}{N_s - 1} \quad \sum_{\vec{s}_{spike}} \vec{s} \\, \vec{s}^T - STA \\; STA^T ,$$

where $N_s$ is the number of spikes recorded.  Each current in the ensemble ($\vec{s}$) is multiplied by it's transpose to form a correlation matrix, the STA covariance is removed and then the sum in normalized by 1 minus the total number of spikes to give an unbiased mean.  By performing the same analysis on the noise (random spike times) ensemble, we can compute the covariance difference matrix ($C_{\Delta}$):

$$C_{\Delta} = C_{spike} - C_{noise}$$

We can now compute the eigenvectors on the $C_{\Delta} matrix and extract the significant eigenvectors that describe the variance in our data.

```python
def findSTCov(STA, STE, STE_noise, dt):
    '''Perform Spike Triggered Covariance analysis on an ensemble of \
    spike triggered stimuli.

    :param STA: a vector containing the Spike Triggered Average.
    :param STE: a matrix containing the Spike Triggered Ensemble.
    :param STE_noise: a matrix containing an ensemble of randomly placed spikes.
    :param dt: sampling rate or time derivative. Should be corrected \
                for any down sampling that has occured.

    :returns: Data_E, Data_Vect
    '''
    # ensure STA proper length:
    STA_len = len(STA) * dt
    End = STA_len / dt
    spike_count = len(STE[:, 0])
    
    # preallocate memory:
    C_spike = np.zeros((End, End))
    C_prior = np.zeros((End, End))
    
    # create STA covariance matrix
    mat_STA = STA * STA[:, np.newaxis] * (spike_count / (spike_count - 1.0))

    for i in range(len(STE)):
        a = STE[i, :]
        b = STE_noise[i, :]
        
        # generate spike and prior Cov matrix for each element:
        mat_spike = a * a[:, np.newaxis]
        mat_prior = b * b[:, np.newaxis]

        # remove STA from each element of STE:
        C_spike += (mat_spike - mat_STA)
        
        # construct the prior (noise) cov matrix
        C_prior += mat_prior

    # Find means of spike and prior cov matrix:
    C_spike /= (spike_count - 1.0)
    C_prior /= (spike_count - 1.0)

    # Find delta cov 
    C_delta = C_spike - C_prior

    # Compute Eigenvalues, Eigenvectors
    vals, vects = np.linalg.eig(C_delta)
    
    # Sort the EigenValues:
    I = np.argsort(vals)
    vals = vals[I]
    # Sort the Eigenvectors
    vects = vects[:,I]

    return vals, vects
```

### VI. Correcting for correlation
***

The correlations introduced into the stimulus violate the assumption of a linear systems analysis that the inputs are derived from a gaussian distribution.  Therefore, we must correct the stimulus in order to properly estimate the linear receptive fields. We follow the approach developed by the Dan group (Touryan et al., 2005).  

The first step in the process is whitening the stimulus.  We can accomplish this by computing the eigen system of the stimulus correlation matrix, $S$.  

$$S_{w} = SU \begin{pmatrix} 
                \frac{1}{\sqrt{\lambda_1}} & & 0 \\\
                 & \ddots & \\\
                 0 & & \frac{1}{\sqrt{\lambda_n}}
            \end{pmatrix} , $$

The resulting $S_{w}$ is a representation of the stimulus in a whitened spaced.  This method is applied to each stimulus in the spike triggered ensemble, forming a whitened ensemble, $\\{S_{w}\\}$.  Now that the stimulus is in a whitened space, we have met the Gaussian requirement of STC and can compute the eigen system as above.  However, the resulting vectors will be in an imaginary whitened space and we would like to have the leading eigenvectors in a space defined in the original current space.  To do this we must move the vectors back into the original space by scaling each eigenvector by the inverse of the square of the eigenvalues ($\lambda_1, \dotso  \lambda_n$) and then rotating the coordinates back into stimulus space through multiplication of inverse of the eigenvectors ($U$): 

$$V = V_{w}^{T}  \begin{pmatrix} 
                \frac{1}{\sqrt{\lambda_1}} & & 0 \\\
                 & \ddots & \\\
                 0 & & \frac{1}{\sqrt{\lambda_n}}
            \end{pmatrix}  U^{-1}, $$

where $V_{w}$ represents the significant eigenvectors in the whitened space.  The resulting $V$ is transformed back into the original stimulus space with the bias introduced by correlations removed.

We begin the normaliziation process below by first defining a function to perform the first equation above, which we call `whiten`.  This function takes a stimulus vector as input, $\vec{s}$.  The correlation coefficients are then computed between $\vec{s}$ and itself at time-1, since the stimuli are correlated in time.  The eigenvalues and eigenvectors of this correlation matrix are found and then matrix multiplication is performed as written.  This moves the entire stimulus into a whitened space.  We then recompute the STE and subsequently the STC analysis.  Finally, we define a function, `unwhiten`, to move back into stimulus space.  

```python
# Whiten a 1D stimulus
def whiten(stim):
    '''Whiten a stimulus using eigenvalue decomposition.
    
    :param stim: a stimulus vector
    :returns: N-1 length whitened stimulus vector
    '''
    # compute the correlation coefficient matrix
    c = np.corrcoef(stim[1:], stim[:-1])
    # compute the eigenvectors
    val, vect = np.linalg.eig(c)
    # reshape and flip the eigenvalues
    val = (np.sqrt(val) ** -1) * np.identity(val.shape[0]) # make diagonal
    # create a 2xN array
    xy = np.array([stim[1:], stim[:-1]]).T
    # perform matrix multiplication
    step1 = np.dot(xy, vect)
    white = np.dot(step1, val)
    return white[:,1] #, val, vect

def unwhiten(stim, eigen):
    '''Move an eigenvector defined in whitened space into stimulus space

    :param stim: original (unwhitened) stimulus vector
    :param eigen: eigenvector to be transformed

    :returns: transformed eigenvector
    '''
    # compute the correlation coefficient matrix
    c = np.corrcoef(stim[1:], stim[:-1])
    # compute the eigenvectors
    val, vect = np.linalg.eig(c)
    # reshape and flip the eigenvalues
    val = (np.sqrt(val) ** -1) * np.identity(val.shape[0]) # make diagonal
    xy = np.array([eigen[1:], eigen[:-1]])
    # perform matrix multiplication
    step1 = np.dot(xy.T, val)
    unwhite = np.dot(step1, vect ** -1)
    return unwhite[:,1] #, val, vect
```

### Eigenvector projection onto stimulus
***

Finally, in order to compare the results of these analyses we project the whitened eigenvectors back onto the ensemble of spike triggered stimuli.  The resulting distribution of lengths gives us a measure of the similarity between the significant eigenvector and the stimulus that lead to spike generation.  For simplicity, we only plot the distribution of projections of the leading eigenvector derived from each noise condition along its respective STE.

```python

def projection(vect, STA, STE, STE_noise):
    '''
    '''
    length = len(STE[:, 0])
    Priorproj = np.zeros((length))
    STAproj = np.zeros((length))
    Eigproj = np.zeros((length))
                       
    for i in range(length):
        Priorproj[i] = np.dot(STA,STE_noise[i, :])
        STAproj[i] = np.dot(STA,STE[i, :])
        Eigproj[i] = np.dot(vect,STE[i, :])
        
    return Priorproj, STAproj, Eigproj

Aprior, Aproj, AEigproj = projection(whiteVect_A[:, 0], STA_A, STE_A, STE_noiseA)
Bprior, Bproj, BEigproj = projection(whiteVect_B[:, 0], STA_B, STE_B, STE_noiseB)
Cprior, Cproj, CEigproj = projection(whiteVect_C[:, 0], STA_C, STE_C, STE_noiseC)

fig = plt.figure(figsize=(16,4))
ax1 = fig.add_subplot(133)
bins = np.arange(-75,75, 5)

ax1.hist(AEigproj, bins=bins, normed=True, alpha = 0.3)
ax1.hist(BEigproj, bins=bins, normed=True, alpha = 0.3)
ax1.hist(CEigproj, bins=bins, normed=True, alpha = 0.3)

TufteAxis(ax1)
TufteAxis(ax2)
TufteAxis(ax3)

ax1.set_ylabel('density')
ax2.set_xlabel('projection')

plt.show()
```
