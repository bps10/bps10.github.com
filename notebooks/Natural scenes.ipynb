```python
from __future__ import division
import numpy as np
import matplotlib.pylab as plt

dt = 0.01 # in msec
tim = 20000 # in msec

def TufteAxis(ax, log=False):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    if not log:
        ax.spines['bottom'].set_position(('outward',10))
        ax.spines['left'].set_position(('outward',10))
```

### I. Constructing stimuli
***

```python
def CorrelatedNoise1D(numSamples, standardDev=0.3, meanValue=0, alphaValue=1, seed=None):
    """
    Produces $\frac{1}{f^\alpha}$ temporally correlated noise.

    :param numSamples: number of samples to generate
    :param standardDev: standard deviation of the noise (default = 0.3)
    :param meanValue: mean value of the noise (default = 0);
    :param alphaValue: type of correlation (i.e., 1 = 1/f, 2 = 1/f^2)

    :returns: y = row vector of the correlated noise
    """

    # Find the next power of two.
    nextP2 = 2 ** np.ceil(np.log2(abs(numSamples)))
    
    # Generate the noise sequence.
    if seed != None:
        np.random.seed(seed)
    x = np.random.rand(nextP2)
    
    # calculate fft and then perform 1/f multiplication
    f = np.fft.fft(x)
    
    # prepare a vector for 1/f^alpha multiplication
    n = np.arange(1, nextP2 / 2 + 2) ** alphaValue

    # It's the power spectrum, so take the square root.
    n = np.sqrt(n) 

    # Multiply the left half of the fft by the power spectrum
    f[:nextP2 / 2 + 1] = f[:nextP2 / 2 + 1] / n
   
    # The right hand of the fft is the mirror image of the left hand
    f_half = f[1:nextP2 / 2]
    f_reverse = f_half[::-1]
    f[nextP2 / 2 + 1:] = np.real(f_reverse) - (1j * np.imag(f_reverse))
   
    y = np.fft.ifft(f) # Take the inverse FFT.
    y = np.real(y[:numSamples]) # Just take the real component.
    y = y.T # Make sure it's a row.
    
    # Make sure it's the right mean and standard deviation.
    y = y / np.std(y) * standardDev
    y = y - np.mean(y) + meanValue

    return y
```

### II. $1 / f^{\alpha}$ noise: manipulating $\alpha$
***


```python
# Generate noise stimuli with various degrees of correlation:
stimA = CorrelatedNoise1D(tim / dt, 5, 0, 0.5, 3)
stimB = CorrelatedNoise1D(tim / dt, 5, 0, 1, 3)
stimC = CorrelatedNoise1D(tim / dt, 5, 0, 2, 3)

# Compute power spectrum:
powerA = abs(np.fft.fftshift(np.real(np.fft.fft(stimA))))
powerB = abs(np.fft.fftshift(np.real(np.fft.fft(stimB))))
powerC = abs(np.fft.fftshift(np.real(np.fft.fft(stimC))))


# Plot stimuli and power spectrum:
fig = plt.figure(figsize=(12,8))
plt.rc('font', **{'size': 16, })

ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax6 = fig.add_subplot(326)

ax1.plot(np.arange(0, 100 / dt) * dt, stimA[0:100 / dt])
ax2.loglog(powerA[len(powerA) / 2 + 2:-1])
ax3.plot(np.arange(0, 100 / dt) * dt, stimB[0:100 / dt])
ax4.loglog(powerB[len(powerB) / 2 + 2:-1])
ax5.plot(np.arange(0, 100 / dt) * dt, stimC[0:100 / dt])
ax6.loglog(powerC[len(powerC) / 2 + 2:-1])


TufteAxis(ax1)
TufteAxis(ax2, True)
TufteAxis(ax3)
TufteAxis(ax4, True)
TufteAxis(ax5)
TufteAxis(ax6, True)

ax1.set_title('$1/f^{\\alpha}$ noise')
ax2.set_title('power spectrum')
ax3.set_ylabel('current (nA)')
ax4.set_ylabel('power')
ax5.set_xlabel('time (s)')
ax6.set_xlabel('frequency (Hz)')

plt.tight_layout()
plt.show()
```

### III. Hodgkin-Huxley model
***


```python
def HHmodel(Current, dt):
    """
    """
    TIME = len(Current) # in msec
    dt = dt # in msec
    
    ## in mS/cm^2
    gK = 36
    gNa = 120
    gL = 0.3
    
    ## in mV
    VK = -12
    VNa = 115
    VL = 10.613
    
    C = 1
    A = np.pi * 30 ** 2 ## in microns^2
    
    ## Resting potential variables
    V = 0 ##mV
    n = 0.0816
    m = 0.1003
    h = 0.0643
    
    ## Functions
    # K channel
    alpha_n = np.vectorize(lambda v: 0.01 * (-v + 10) / 
                (np.exp((-v + 10) / 10) - 1) if v != 10 else 0.1)
    beta_n  = lambda v: 0.125 * np.exp(-v / 80)
    
    # Na channel (activating)
    alpha_m = np.vectorize(lambda v: 0.1 * (-v + 25) / 
                (np.exp((-v + 25) / 10) - 1) if v != 25 else 1)
    beta_m  = lambda v: 4 * np.exp(-v / 18)
    
    # Na channel (inactivating)
    alpha_h = lambda v: 0.07 * np.exp(-v / 20)
    beta_h  = lambda v: 1 / (np.exp((-v + 30) / 10) + 1)
    
    
    ## Records
    Voltage = np.zeros((TIME))
    Voltage[0] = V
    
    # Euler integration of the equations
    for i in range(0, TIME):
    
        I = Current[i]
        
        m += dt * (alpha_m(V) * (1 - m) - beta_m(V) * m)
        h += dt * (alpha_h(V) * (1 - h) - beta_h(V) * h)
        n += dt * (alpha_n(V) * (1 - n) - beta_n(V) * n)
    
        V = V + ( (I) - ((gK * (n ** 4))*(V - VK)) - ((gNa * (m ** 3)) 
                * h * (V - VNa)) - (gL * (V - VL))) * dt
        
        Voltage[i] = V
    
    return Voltage
    
```

####Plot figures

```python
VoltageA = HHmodel(stimA, dt)
VoltageB = HHmodel(stimB, dt)
VoltageC = HHmodel(stimC, dt)

fig = plt.figure(figsize=(7,10))
plt.rc('font', **{'size': 16, })
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)

ax1.plot(np.arange(0, 1000 / dt) * dt + 200, VoltageA[200 / dt:1200 / dt])
ax2.plot(np.arange(0, 1000 / dt) * dt + 200, VoltageB[200 / dt:1200 / dt])
ax3.plot(np.arange(0, 1000 / dt) * dt + 200, VoltageC[200 / dt:1200 / dt])


TufteAxis(ax1)
TufteAxis(ax2)
TufteAxis(ax3)

ax2.set_ylabel('membrane potential (mV)')
ax3.set_xlabel('time (s)')


plt.tight_layout()
plt.show()
```

### IV. STA
***

```python
def runs(bits):
    '''
    '''
    # make sure all runs of ones are well-bounded
    bounded = np.hstack(([0], bits, [0]))
    # get 1 at run starts and -1 at run ends
    difs = np.diff(bounded)
    run_starts, = np.where(difs > 0)
    run_ends, = np.where(difs < 0)
    # eliminate noise or short duration chirps
    length = run_ends-run_starts
    return run_starts, run_ends

## Find Spikes while ignoring those that fall too close to the beginning or end
def findSpikes(Voltage, dt):
    '''
    '''
    Spikes = Voltage > 20
    Spikes = runs(Spikes)
    S_beg = np.where(Spikes[0] > int(40 / dt * 2))
    S_end = np.where(Spikes[0] < (len(Voltage) - (int(40 / dt * 2) + 20)))
    Length = np.arange(min(S_beg[0]),max(S_end[0]) + 1)
    Spikes = Spikes[0][Length], Spikes[1][Length]
    
    return Spikes

# Find spike triggered ensamble
def findSTE(current, Voltage, Spikes, dt, STA_len=30):
    '''
    '''
    # Preallocate memory for ensamble of spike triggered current:
    STE = np.zeros((len(Spikes[0]), int(STA_len / dt)))
    STE_noise = np.zeros((len(Spikes[0]), int(STA_len / dt)))
    
    # Extract current preceeding spikes
    for i in range(0,len(Spikes[0])):
                   
        Peak = np.arange(Spikes[0][i],Spikes[1][i])
        Peak = Voltage[Peak]
        Height = np.argmax(Peak)
        
        Loc = Height + Spikes[0][i]
        RandLoc = np.random.random_integers(STA_len / dt, 
                                            (len(current) - (STA_len / dt)))
        
        Range = np.arange(Loc - (STA_len / dt), Loc, dtype=int)
        RandRange = np.arange(RandLoc - (STA_len / dt), RandLoc, dtype=int)
        
        STE[i, :] = current[Range]
        STE_noise[i, :] = current[RandRange]
    
    return STE, STE_noise

## firing rate:
def findFiringRate(curr_Record, tim):
    '''Returns firing rate in spikes/sec.

    :param curr_Record: matrix with current preceeding each spike in rows
    :param tim: time of entire epoch in seconds

    :returns: spike count and spike rate (spikes / second)
    '''
    return len(curr_Record[:, 0]), len(curr_Record[:, 0]) / tim
```

####Compute STA:

```python
STA_len = 40 # msec

spikesA = findSpikes(VoltageA, dt)
spikesB = findSpikes(VoltageB, dt)
spikesC = findSpikes(VoltageC, dt)

STE_A, STE_noiseA = findSTE(stimA, VoltageA, spikesA, dt, STA_len)
STE_B, STE_noiseB = findSTE(stimB, VoltageB, spikesB, dt, STA_len)
STE_C, STE_noiseC = findSTE(stimC, VoltageC, spikesC, dt, STA_len)

spike_countA, spikeRateA = findFiringRate(STE_A, tim / 1000) # divide by 1000 to get sec
spike_countB, spikeRateB = findFiringRate(STE_B, tim / 1000)
spike_countC, spikeRateC = findFiringRate(STE_C, tim / 1000)

print 'spikes / sec'
print '-------------------'
print 'alpha: 0.5:', spikeRateA
print 'alpha: 1.0:', spikeRateB
print 'alpha: 2.0:', spikeRateC 

STA_A = STE_A.mean(0)
STA_B = STE_B.mean(0)
STA_C = STE_C.mean(0)
```

#### Plotting the STA

```python
t = np.arange(-STA_len / dt, 0) * dt

# plot STA
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)

ax1.plot(t, STA_A, linewidth=2, color='k')
ax1.fill_between(t, STA_A + STE_A.std(0), 
                 STA_A - STE_A.std(0),
                 facecolor='gray', alpha=0.8, edgecolor='gray')

ax2.plot(t, STA_B, linewidth=2, color='k')
ax2.fill_between(t, STA_B + STE_B.std(0), 
                 STA_B - STE_B.std(0),
                 facecolor='gray', alpha=0.8, edgecolor='gray')

ax3.plot(t, STA_C, linewidth=2, color='k')
ax3.fill_between(t,STA_C + STE_C.std(0), 
                 STA_C - STE_C.std(0),
                 facecolor='gray', alpha=0.8, edgecolor='gray')

TufteAxis(ax1)
TufteAxis(ax2)
TufteAxis(ax3)

ax2.set_ylabel('current (nA)')
ax3.set_xlabel('time (msec) before spike')

plt.tight_layout()
plt.show()
```

### V. Spike triggered covariance
***


```python
def findSTCov(STA, STE, stimA, STE_noise, spike_count, STA_len, dt):
   '''
   '''
   End = STA_len / dt

   # preallocate memory:
   Data_C_spike = np.zeros((End, End))
   C_prior = np.zeros((End, End))

   # ensure STA proper length:
   STA = STA[0:End]
    
   # create STA covariance matrix
   mat_STA = STA * STA[:, np.newaxis] * spike_count / (spike_count - 1.0)

   for i in range(len(STE)):
       a = STE[i, :End]
       b = STE_noise[i, :End]

       mat_spike = a * a[:, np.newaxis]
       mat_prior = b * b[:, np.newaxis]

       # remove STA from each element of STE:
       Data_C_spike += (mat_spike - mat_STA)
        
       C_prior += mat_prior

   # Find means
   Data_C_spike = Data_C_spike / (spike_count - 1.0)
   C_prior = C_prior / (spike_count - 1.0)

   # Find delta cov 
   Data_C_delta = Data_C_spike - C_prior

   # EigenValues, EigenVectors
   Data_E, V = np.linalg.eig(Data_C_delta)
   I = np.argsort(Data_E)
   Data_E = Data_E[I][:( STA_len / dt)]
   Data_Vect = V[:,I][:, :( STA_len / dt)]
    
   return Data_E, Data_Vect
```

####Compute Spike triggered covariance

```python
EigVal_A, EigVect_A = findSTCov(STA_A, STE_A, stimA, STE_noiseA, spike_countA, STA_len, dt)
EigVal_B, EigVect_B = findSTCov(STA_B, STE_B, stimB, STE_noiseB, spike_countB, STA_len, dt)
EigVal_C, EigVect_C = findSTCov(STA_C, STE_C, stimC, STE_noiseC, spike_countC, STA_len, dt)
```

```python
# plot Eigen value spectrum:
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax6 = fig.add_subplot(326)

ax1.plot(EigVal_A, 'o')
ax2.plot(-1 * EigVect_A[:, -1], '-', linewidth=2)
#ax2.plot(-1 * EigVect_A[:, -2], 'r-', linewidth=2)

ax3.plot(EigVal_B, 'o')
ax4.plot(-1 * EigVect_B[:, -1], '-', linewidth=2)
#ax4.plot(-1 * EigVect_A[:, -2], 'r-', linewidth=2)

ax5.plot(EigVal_C, 'o')
ax6.plot(-1 * EigVect_C[:, -1], '-', linewidth=2)
#ax6.plot(-1 * EigVect_C[:, 0], 'r-', linewidth=2)

TufteAxis(ax1)
TufteAxis(ax2)
TufteAxis(ax3)
TufteAxis(ax4)
TufteAxis(ax5)
TufteAxis(ax6)

ax2.set_ylabel('eigen value')
ax3.set_xlabel('rank')

plt.tight_layout()
plt.show()
```
