### I. Constructing stimuli
***

The goal of this project was to simulate the effects of introducing correlation into Gaussian stimuli during noise analysis.  The function defined below, `CorrelatedNoise1D`, introduces correlation to a sequence of random numbers in the Fourier domain.  

The steps taken are as follows:

1. The function begins by finding the next power of two, which will speed up computations in the Fourier transform later. 
2. The noise sequence is then generated using a Twister-Mercene algorithm.  As an optional parameter, the user can specify a seed for the algorithm that can be useful when regenerating the stimuli later or for the purposes of comparison.  Here, we pass in a seed in order to directly compare the effect of correlation on neuronal responses.  Using the exact same random sequence for each stimulus will eliminate random variability due to finite sampling.   
3. The noise sequence is then moved into the Fourier domain, where convolution between a signal and a filter (i.e. $1/f^{\alpha}$) can be computed much more efficiently through multiplication.  
4. We then generate a vector, n, with the $1/f^{\alpha}$ properties specified by the parameter `alphaValue`.  An $\alpha = 0$ will not introduce any correlation (anything raised to 0 equals 1), while increasing $\alpha$ increases the correlation in the signal.  Since we are manipulating the power spectrum, we take the square root of n, as we will multiply the inverse by the Fourier components.
5. The left hand of the Fourier spectrum is multiplied by the power spectrum. Since the Fourier spectrum is symmetric, we flip the power spectrum, n, and multiply the second half.
6. The inverse Fourier transform is than taken to move the signal back into the time domain and we select the first half of the signal (since it is symmetric).
7. Finally, the signal is corrected to have the correct mean and standard deviation.


```python
def CorrelatedNoise1D(numSamples, standardDev=0.3, meanValue=0, alphaValue=1, seed=None):
    """
    Produces $\frac{1}{f^\alpha}$ temporally correlated noise.

    :param numSamples: number of samples to generate
    :param standardDev: standard deviation of the noise (default = 0.3)
    :param meanValue: mean value of the noise (default = 0);
    :param alphaValue: type of correlation (i.e., 1 = 1/f, 2 = 1/f^2)

    :returns: y = row vector of the correlated noise
    """

    # 1. Find the next power of two.
    nextP2 = 2 ** np.ceil(np.log2(abs(numSamples)))
    
    # 2. Generate the noise sequence.
    if seed != None:
        np.random.seed(seed)
    x = np.random.rand(nextP2)
    
    # 3. calculate fft and then perform 1/f multiplication
    f = np.fft.fft(x)
    
    # 4. prepare a vector for 1/f^alpha multiplication
    n = np.arange(1, nextP2 / 2 + 2) ** alphaValue
    # It's the power spectrum, so take the square root.
    n = np.sqrt(n) 

    # 5. Multiply the left half of the fft by the power spectrum
    f[:nextP2 / 2 + 1] = f[:nextP2 / 2 + 1] / n
    f[nextP2 / 2 - 1:] = f[nextP2 / 2 - 1:] / n[::-1]
   
    # 6. Take the inverse FFT.
    y = np.fft.ifft(f) # 
    y = np.real(y[:numSamples]) # Just take the real component.
    y = y.T # Make sure it's a row.
    
    # 7. Make sure it's the right mean and standard deviation.
    y = y / np.std(y) * standardDev
    y = y - np.mean(y) + meanValue

    return y
```

### III. Hodgkin-Huxley model
***

In 1952, Hodgkin and Huxley came up with a set of equations that described the initiation and propagation of an action potential. The primary equation, describing a neuron's membrane potential, is as follows: 

$$C_{m} \frac{dV_{m}}{dt} = I_{ext} - g_{K}n^4(V_{m}-E_{K}) - g_{Na}m^3h(V_{m}-ENa) - g_{l}(V_{m}-E_{l})$$

$n$, $m$, and $h$ are the gating variable that correspond to the probability that a channel is open, and are described by a set of differential equations,

$$\frac{dm}{dt} = \alpha_{m}(V)(1-m)-\beta_{m}(V)m$$
<br>
$$\frac{dn}{dt} = \alpha_{n}(V)(1-n)-\beta_{n}(V)n$$
<br>
$$\frac{dh}{dt} = \alpha_{h}(V)(1-h)-\beta_{h}(V)h$$

The rate constants $\alpha$ and $\beta$ of each equation, are themselves described by a set of voltage dependent differential equations. Below, we use these equations, as well as the resting (initial) values, to simulate a neuron firing an action potential.

We adopt here the parameters used by Agüera y Arcas et al. (2003).

```python
def HHmodel(Current, dt):
    """Integrates the Hodgkin Huxley equations.
    The current parameters of the model are taken from 
    """
    TIME = len(Current) # in msec
    dt = dt # in msec
    
    # Define the cell's initial conditions
    # Conductances (in mS/cm^2):
    gK = 36
    gNa = 120
    gL = 0.3
    
    # Define the equilibrium potentials and conductances for sodium, potassium,
    # and the leak current
    VK = -12
    VNa = 115
    VL = 10.613

    V = 0 ##mV
    n = 0.0816
    m = 0.1003
    h = 0.0643
    
    # Functions to Calculate alphas and betas for the intial membrane voltage (based on
    #the equations described by Hodgkin and Huxley, modified to refelct the
    # initial Vm)
    
    # K channel
    alpha_n = np.vectorize(lambda v: 0.01 * (-v + 10) / 
                (np.exp((-v + 10) / 10) - 1) if v != 10 else 0.1)
    beta_n  = lambda v: 0.125 * np.exp(-v / 80)
    
    # Na channel (activating)
    alpha_m = np.vectorize(lambda v: 0.1 * (-v + 25) / 
                (np.exp((-v + 25) / 10) - 1) if v != 25 else 1)
    beta_m  = lambda v: 4 * np.exp(-v / 18)
    
    # Na channel (inactivating)
    alpha_h = lambda v: 0.07 * np.exp(-v / 20)
    beta_h  = lambda v: 1 / (np.exp((-v + 30) / 10) + 1)
    
    
    ## Preallocate for speed:
    Voltage = np.zeros((TIME))
    Voltage[0] = V
    
    # Euler integration of the equations
    for i in range(0, TIME):
    
        I = Current[i]
        
        # The equations that describe the gating variables:
        m += dt * (alpha_m(V) * (1 - m) - beta_m(V) * m)
        h += dt * (alpha_h(V) * (1 - h) - beta_h(V) * h)
        n += dt * (alpha_n(V) * (1 - n) - beta_n(V) * n)
    
        # The equation describing the cell's net ionic current, rearranged to
        # solve for Vm.
        V = V + ( (I) - ((gK * (n ** 4))*(V - VK)) - ((gNa * (m ** 3)) 
                * h * (V - VNa)) - (gL * (V - VL))) * dt
        
        Voltage[i] = V
    
    return Voltage
    
```

### IV. Spike Triggered Average (STA)
***

The first step in computing the STA of a neuron is to find the locations of spikes in the data.  Since we are simulating the voltage trace (equivalent to whole cell recording), the detection of spikes is very simple. The function called `findSpikes` below efficiently picks out the spikes with a threshold method:

1. The voltage trace is thresholded at 20mV.  That is, values above 20mV will equal 1, while those less than or equal to 20 will become 0.
2. The thresholded data is then passed to a function called runs which finds the beginning and end of 'runs' of 1s.  This gives us the location of the threshold crossing at the rising and falling phase of the spike.
3. Spike times are checked to ensure that spikes falling too near the beginning or end of the trail are ignored (otherwise they would mess up the STA).

Now having a list of the spike locations, we can extract the current preceeding each spike and place it into a matrix which we will call the Spike Triggered Ensemble (STE).  The function `findSTE` will accomplish this task.  This function will extract the current preceeding each spike, as well as an equal number of randomly generated spikes that will form the prior distribution (i.e. noise).  The algorithm for this proceedure is actually as straightforward as it sounds, as indicated in the code comments.

```python
def runs(bits):
    '''Find zero crossing points from a boolean array.

    :param bits: boolean array.
    '''
    # make sure all runs of ones are well-bounded
    bounded = np.hstack(([0], bits, [0]))
    # get 1 at run starts and -1 at run ends
    difs = np.diff(bounded)
    run_starts, = np.where(difs > 0)
    run_ends, = np.where(difs < 0)

    return run_starts, run_ends

def findSpikes(Voltage, dt):
    '''Find Spikes while ignoring those that fall too close \
    to the beginning or end.

    :param Voltage: voltage trace.
    :param dt: time step used during sampling.

    :returns: Two arrays. One containing location of spike start times \
    the other constainig the spike end times.
    '''
    # 1. Threshold = 20mV
    Spikes = Voltage > 20 
    
    # 2. Find beginning and end of spikes
    Spikes = runs(Spikes)
    
    # 3. Make sure spikes don't fall too close to beginning or end
    # of the trial. 
    S_beg = np.where(Spikes[0] > int(40 / dt * 2))
    S_end = np.where(Spikes[0] < (len(Voltage) - (int(40 / dt * 2) + 20)))
    # create an index of spike times
    Length = np.arange(min(S_beg[0]),max(S_end[0]) + 1)
    Spikes = Spikes[0][Length], Spikes[1][Length]
    
    return Spikes

def findSTE(current, Voltage, Spikes, dt, STA_len=40, downS=5):
    '''Find spike triggered ensamble.

    :param current: stimulus played to neuron.
    :param Voltage: voltage trace.
    :param Spikes: a list of spike locations (beginning, end). 
    :param dt: time step used.
    :param STA_len: length of the STA (in msec).

    :returns: Spike triggered ensemble and a randomly generated \
    ensemble.
    '''
    # Preallocate memory for ensamble of spike triggered current:
    STE = np.zeros((len(Spikes[0]), int(STA_len / dt / downS)))
    STE_noise = np.zeros((len(Spikes[0]), int(STA_len / dt / downS)))
    
    # 1. Extract current preceeding spikes:
    for i in range(0,len(Spikes[0])):
        
        # 2. Extract current preceeding spikes:
        Peak = np.arange(Spikes[0][i],Spikes[1][i])
        Peak = Voltage[Peak]
        Height = np.argmax(Peak)
        
        # 3. Find spike location and range:
        Loc = Height + Spikes[0][i]
        # Find a random location from the stimulus:
        RandLoc = np.random.random_integers(STA_len / dt, 
                                            (len(current) - (STA_len / dt)))
        
        Range = np.arange(Loc - (STA_len / dt), Loc, dtype=int)
        RandRange = np.arange(RandLoc - (STA_len / dt), RandLoc, dtype=int)
        
        # 4. Extract currents and downsample if required:
        cur = current[Range]
        STE[i, :] = cur[::downS]
        cur = current[RandRange]
        STE_noise[i, :] = cur[::downS]
    
    return STE, STE_noise

def findFiringRate(curr_Record, tim):
    '''Returns firing rate in spikes/sec.

    :param curr_Record: matrix with current preceeding each spike in rows
    :param tim: time of entire epoch in seconds

    :returns: spike count and spike rate (spikes / second)
    '''
    return len(curr_Record[:, 0]), len(curr_Record[:, 0]) / tim
```

####Compute STA:

Now we find spikes and compile the STE using the above defined functions.  Finally, we print the spike rate for each alpha value and compute the STA, simply taking the mean of across the STE at each sampling position.

```python
spikesA = findSpikes(VoltageA, dt)
spikesB = findSpikes(VoltageB, dt)
spikesC = findSpikes(VoltageC, dt)

STE_A, STE_noiseA = findSTE(stimA, VoltageA, spikesA, dt, STA_len, downS)
STE_B, STE_noiseB = findSTE(stimB, VoltageB, spikesB, dt, STA_len, downS)
STE_C, STE_noiseC = findSTE(stimC, VoltageC, spikesC, dt, STA_len, downS)

spike_countA, spikeRateA = findFiringRate(STE_A, tim / 1000) # divide by 1000 to get sec
spike_countB, spikeRateB = findFiringRate(STE_B, tim / 1000)
spike_countC, spikeRateC = findFiringRate(STE_C, tim / 1000)

print 'spikes / sec'
print '-------------------'
print 'alpha 0.5:', spikeRateA
print 'alpha 1.0:', spikeRateB
print 'alpha 2.0:', spikeRateC 

STA_A = STE_A.mean(0)
STA_B = STE_B.mean(0)
STA_C = STE_C.mean(0)
```

### V. Spike triggered covariance
***

$$C_{m,n} = \frac{1}{N} \sum_{i=1}^{N} S_{m}S_{n}$$

```python
def findSTCov(STA, STE, STE_noise, dt):
    '''Perform Spike Triggered Covariance analysis on an ensemble of \
    spike triggered stimuli.

    :param STA: a vector containing the Spike Triggered Average.
    :param STE: a matrix containing the Spike Triggered Ensemble.
    :param STE_noise: a matrix containing an ensemble of randomly placed spikes.
    :param dt: sampling rate or time derivative. Should be corrected \
                for any down sampling that has occured.

    :returns: Data_E, Data_Vect
    '''
    # ensure STA proper length:
    STA_len = len(STA) * dt
    
    End = STA_len / dt
    
    spike_count = len(STE[:, 0])
    # preallocate memory:
    Data_C_spike = np.zeros((End, End))
    C_prior = np.zeros((End, End))
    
    # create STA covariance matrix
    mat_STA = STA * STA[:, np.newaxis] * spike_count / (spike_count - 1.0)

    for i in range(len(STE)):
        a = STE[i, :]
        b = STE_noise[i, :]

        mat_spike = a * a[:, np.newaxis]
        mat_prior = b * b[:, np.newaxis]

        # remove STA from each element of STE:
        Data_C_spike += (mat_spike - mat_STA)
        
        C_prior += mat_prior

    # Find means
    Data_C_spike = Data_C_spike / (spike_count - 1.0)
    C_prior = C_prior / (spike_count - 1.0)

    # Find delta cov 
    Data_C_delta = Data_C_spike - C_prior

    # EigenValues, EigenVectors
    Data_E, V = np.linalg.eig(Data_C_delta)
    I = np.argsort(Data_E)
    Data_E = Data_E[I][:( STA_len / dt)]
    Data_Vect = V[:,I][:, :( STA_len / dt)]
    
    return Data_E, Data_Vect
```

####Compute Spike triggered covariance

```python
# Compute the Eigen values and Eigen vectors for each stimulus/response pair:
EigVal_A, EigVect_A = findSTCov(STA_A, STE_A, STE_noiseA, dt / downS)
EigVal_B, EigVect_B = findSTCov(STA_B, STE_B, STE_noiseB, dt / downS)
EigVal_C, EigVect_C = findSTCov(STA_C, STE_C, STE_noiseC, dt / downS)
```

```python
# plot Eigen value spectrum:
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax6 = fig.add_subplot(326)

x_ = np.arange(-STA_len / dt / downS, 0) * dt * downS
ax1.plot(EigVal_A / np.sum(EigVal_A), 'o')
ax2.plot(x_, EigVect_A[:, -1], '-', linewidth=2)
#ax2.plot(-1 * EigVect_A[:, -2], 'r-', linewidth=2)

ax3.plot(EigVal_B / np.sum(EigVal_B), 'o')
ax4.plot(x_, EigVect_B[:, -1], '-', linewidth=2)
#ax4.plot(x_, -1 * EigVect_A[:, -2], 'r-', linewidth=2)

ax5.plot(EigVal_C / np.sum(EigVal_C), 'o')
ax6.plot(x_, EigVect_C[:, -1], '-', linewidth=2)
#ax6.plot(-1 * EigVect_C[:, 0], 'r-', linewidth=2)

TufteAxis(ax1)
ax1.set_xlim([-10, len(EigVal_A) + 10])
TufteAxis(ax2)
TufteAxis(ax3)
ax3.set_xlim([-10, len(EigVal_B) + 10])
TufteAxis(ax4)
TufteAxis(ax5)
ax5.set_xlim([-10, len(EigVal_C) + 10])
TufteAxis(ax6)

ax3.set_ylabel('normalized eigen value')
ax4.set_ylabel('pA')
ax5.set_xlabel('rank')
ax6.set_xlabel('time (msec) before spike')


plt.tight_layout()
plt.show()
```

### VI. Correcting for correlation
***

The correlations introduced into the stimulus violates an assumption of a linear systems analysis, namely that the inputs are derived from a gaussian distribution.  Therefore, we must correct the stimulus in order to properly estimate the linear receptive fields.

We follow the approach developed by the Dan group (Touryan et al., 2005).  

$$S_{w} = SU \begin{pmatrix} 
                \frac{1}{\sqrt{\lambda_1}} & & 0 \\\
                 & \ddots & \\\
                 0 & & \frac{1}{\sqrt{\lambda_n}}
            \end{pmatrix} , $$


$$V = V_{w}^{T}  \begin{pmatrix} 
                \frac{1}{\sqrt{\lambda_1}} & & 0 \\\
                 & \ddots & \\\
                 0 & & \frac{1}{\sqrt{\lambda_n}}
            \end{pmatrix}  U^{-1}, $$

```python
def correctedSTCov(STA, STE, STE_noise, dt):
    '''Perform Spike Triggered Covariance analysis on an ensemble of \
    spike triggered stimuli.

    :param STA: a vector containing the Spike Triggered Average.
    :param STE: a matrix containing the Spike Triggered Ensemble.
    :param STE_noise: a matrix containing an ensemble of randomly placed spikes.
    :param dt: sampling rate or time derivative. Should be corrected \
                for any down sampling that has occured.

    :returns: Data_E, Data_Vect
    '''
    # ensure STA proper length:
    STA_len = len(STA) * dt
    
    End = STA_len / dt
    
    spike_count = len(STE[:, 0])
    # preallocate memory:
    Data_C_spike = np.zeros((End, End))
    C_prior = np.zeros((End, End))
    
    # create STA covariance matrix
    mat_STA = STA * STA[:, np.newaxis] * spike_count / (spike_count - 1.0)

    for i in range(len(STE)):

        # first remove correlation:\n",
        s = STE[i, :]

        s_mat = s * s[:, np.newaxis]

        s_Val, s_Vect = np.linalg.eig(s_mat)

        temp = np.dot(s, s_Vect)
        whiteSTE = np.dot(temp, (np.sqrt(s_Val) ** -1) * np.identity(s_Val.shape[0]))

        # Then compute cov matrix
        a = whiteSTE
        b = STE_noise[i, :]

        mat_spike = a * a[:, np.newaxis]
        mat_prior = b * b[:, np.newaxis]

        # remove STA from each element of STE:
        Data_C_spike += (mat_spike - mat_STA)

        C_prior += mat_prior

    # Find means
    Data_C_spike = Data_C_spike / (spike_count - 1.0)
    C_prior = C_prior / (spike_count - 1.0)

    # Find delta cov 
    Data_C_delta = Data_C_spike - C_prior

    # EigenValues, EigenVectors
    Data_E, V = np.linalg.eig(Data_C_delta)
    I = np.argsort(Data_E)
    Data_E = Data_E[I][:( STA_len / dt)]
    Data_Vect = V[:,I][:, :( STA_len / dt)]

    return Data_E, Data_Vect
```

```python
def correctedSTCov(STA, STE, STE_noise, dt):
    '''
    '''
    # ensure STA proper length:
    STA_len = len(STA) * dt
    
    End = STA_len / dt
    
    spike_count = len(STE[:, 0])

    # preallocate memory:
    Data_spike = np.zeros((End, End))
    prior = np.zeros((End, End))
    whiteSTE = np.zeros((End, End))
    n_mat = np.zeros((End, End))
    s_mat = np.zeros((End, End))    
    
    # create STA covariance matrix
    mat_STA = STA * STA[:, np.newaxis] * spike_count / (spike_count - 1.0)
    
    for i in range(len(STE)):
        # first remove correlation:
        s = STE[i, :]
        n = STE_noise[i, :]
        
        s_mat += s * s[:, np.newaxis]
        n_mat += n * n[:, np.newaxis]
        
    s_Val, s_Vect = np.linalg.eig(s_mat)
    n_Val, n_Vect = np.linalg.eig(n_mat)
    p_Val, p_Vect = np.linalg.eig(mat_STA)
        
    temp = np.dot(s_mat, s_Vect)
    whiteSTE = np.dot(temp, (np.sqrt(s_Val) ** -1) * np.identity(s_Val.shape[0]))

    temp = np.dot(n_mat, n_Vect)
    whiteSTEnoise = np.dot(temp, (np.sqrt(n_Val) ** -1) * np.identity(n_Val.shape[0]))

    temp = np.dot(mat_STA, p_Vect)
    prior = np.dot(temp, (np.sqrt(p_Val) ** -1) * np.identity(p_Val.shape[0]))
    
    # remove STA from each element of STE:
    Data_spike += (whiteSTE - whiteSTEnoise)

    # Find means
    Data_spike = Data_spike / (spike_count - 1.0)
    prior = prior / (spike_count - 1.0)

    # Find delta cov 
    Data_delta = Data_spike - prior

    # EigenValues, EigenVectors
    Data_Val, Data_Vect = np.linalg.eig(Data_delta)
    
    temp = np.dot(Data_Vect.T, (np.sqrt(s_Val) ** -1) * np.identity(s_Val.shape[0]))
    EigVect = np.dot(temp, np.linalg.inv(s_Vect))
    
    I = np.argsort(Data_Val)
    Data_Val = Data_Val[I][:(STA_len / dt)]
    EigVect = EigVect[:,I][:, :( STA_len / dt)]
    
    return Data_Val, EigVect
```
